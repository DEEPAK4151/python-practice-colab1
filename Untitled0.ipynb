```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
paragraph="""Green tea has been enjoyed for centuries in various cultures and is known for its numerous health benefits. It is rich in antioxidants called catechins, which are believed to help prevent cell damage and reduce the risk of chronic diseases such as heart disease, diabetes, and certain types of cancer.  Green tea is also known to boost metabolism and aid in weight loss, as well as improve brain function and promote relaxation. Additionally, it contains caffeine and other compounds that can enhance physical performance and endurance. With all these benefits, its no wonder why green tea continues to be a popular and healthy beverage choice for many."""
```

```python
sentences=nltk.sent_tokenize(paragraph)
stemmer=PorterStemmer()
nltk.download('stopwords')
for i in range(len(sentences)):
  words=nltk.word_tokenize(sentences[i])
  words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  sentences=' '.join(words)
```
